```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
print('1 Dimension', arr1, sep='\n', end='\n\n')
```

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print('2 Dimension', arr2, sep = '\n', end = '\n\n')
```

```python
arr3 = np.array([[[1, 2], [3, 4]], [[4, 5], [6, 7]]])
print('Multi Dimension', arr3, sep = '\n')
```

```python
zeros = np.zeros((5, 5))
print(zeros)
```

```python
ones = np.ones((5, 5))
print(ones)
```

```python
identity = np.eye(5)
print(identity)
```

```python
lin_space = np.linspace(1, 50, 2)
print(lin_space)
```

```python
rand_int = np.random.randint(2, 22)
print(rand_int)
```

```python
rand_float = np.random.rand(3, 3)
print(rand_float)
```

```python
a = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
flattened = a.flatten()
print(flattened)
```

```python
arr = np.array([10, 20, 30 ,40, 50])
print(arr[0])
print(arr[-1])
print(arr[1:4])
print(arr[:3])
print(arr[::2])
```

```python
matrix = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print(matrix[0:2])
print()
print(matrix[0:3, 0])
print()
print(matrix[0:2, 0:2])
```

```python
arr = np.array([4, 9, 16, 25, 36])
print(arr + 1)
print(arr * 2)
print(arr ** 2)
print(np.sqrt(arr))
```

```python
print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))
```

```python
arr = np.array([1, 2, 3, 4, 5])
print(np.min(arr))
print(np.max(arr))
print(np.cumsum(arr))
print(np.cumprod(arr))
```
