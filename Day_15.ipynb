```python
n = int(input("Enter a number: "))
c = 0
for i in range(2, (n//2) + 1):
  if ((n%i) == 0):
    c += 1
print("Prime number" if c==0 else "Not a Prime number")
```

```python
def isprime(n):
  for i in range(2, (n//2) + 1):
    if ((n%i) == 0):
      return False
  return True

n = int(input("Enter the number: "))
print("Prime number" if isprime(n) else "Not a Prime number")
```
