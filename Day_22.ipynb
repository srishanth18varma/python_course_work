```python
try:
  n = 10
  print(n)
  print(13 + 12)
  print(int(input('Enter the int: ')))

  d = {1:9, 2:3, 4:7}
  print(d[2])

  l = [34, 456, 134, 3456]
  print(l[10])
  print(l / 9)

except (NameError, TypeError, ValueError, KeyError, IndexError, ZeroDivisionError) as e:
  print('Error Occured: ', e)

else:
  print('No Errorss Occured')

finally:
  print('End of the program')
```
