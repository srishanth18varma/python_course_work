```python
import random
l = ['Python', 'C++', 'Java', 'html', 'C']
```

```python
print('Random number: ', random.random())
print("Random int bw 1 and 3: ", random.randint(1, 3))  #Gives a random integer value.
print('Random number bw 1 and 3: ', random.uniform(1, 3)) #Gives a random number(including decimals).
print('Random str from list L: ', random.choice(l))
print('2 random strs from list L: ', random.choices(l, k = 2))

print('Before shuffling: ', l)
random.shuffle(l)
print('After shuffling: ', l)
```

```python
import collections
s = 'python programming'
n = [1, 2, 3, 4, 5, 6, 1, 2, 3, 4, 2, 1, 1, 1, 1, 1, 2, 3, 4, 5]
```

```python
#Normal code to get number of types an alphabet is repeated in a given string:
d = {}
for i in s:
  if i in d:
    d[i] += 1
  else:
    d[i] = 1

print('Nrml Method: ', d)
print()

#Same case, using the collections module:
print(collections.Counter(s))
```

```python
d = collections.defaultdict(int)
for i in s:
  d[i] = 1
print(d)
```

```python
from datetime import date, time, datetime, timedelta
```

```python
t = date.today()

print('Full Date:', t)
print('Year:', t.year)
print('Month:', t.month)
print('Day:', t.day)
print('Weekday:', t.weekday())
```

```python
n = datetime.now()

print(n)
print(n.year)
print(n.time())
```

```python
#Formating the date and time
print(n.strftime('%Y-%m-%d'))
print(n.strftime('%Y/%m/%d'))
print(n.strftime('%Y/%m/%d %H:%M:%S'))
print(n.strftime('%Y/%m/%d %I:%M:%S %P'))
print(n.strftime('%a %Y/%m/%d %H:%M:%S %P'))
print(n.strftime('%A %d %b %Y %H:%M:%S %P'))
print(n.strftime('%A %d %B %Y %H:%M:%S %P'))
```
