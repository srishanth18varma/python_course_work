# Inheritance

```python
#SINGLE INHERITANCE - one parent, one child
class A:
  def printa(self):
    print("Parent class - A")

class B(A):
  def printb(self):
    print("Child class - B")

a = A()
a.printa()

print()

b = B()
b.printb()
b.printa()
```

```python
#HIERARCHICAL INHERITANCE - one parent, many children
class A:
  def printa(self):
    print("Parent class - A")

class B(A):
  def printb(self):
    print("Child class - B")

class C(A):
  def printc(self):
    print("Child class - C")

class D(A):
  def printd(self):
    print("Child class - D")

a = A()
a.printa()
print()

b = B()
b.printb()
b.printa()
print()

c = C()
c.printc()
c.printa()
print()

d = D()
d.printd()
d.printa()
```

```python
#HIERARCHICAL ORDER - one parent, many children
class A:
  def printa(self):
    print("Parent class - A")

class B(A):
  def printb(self):
    print("Child class - B")

class C(B):
  def printc(self):
    print("Child class - C")

c = C()
c.printa()
c.printb()
c.printc()
```

```python
#HYBRID INHERITANCE - many parents, many children
class A:
  def printa(self):
    print("Parent class - A")

class B:
  def printb(self):
    print("Parent class - B")

class C(B, A):
  def printc(self):
    print("Child class - C")

class D(C):
  def printd(self):
    print("Child class - D")

d = D()
d.printa()
d.printb()
d.printc()
d.printd()
```

```python
class A:
  def print(self):
    print('Class - A')

class B(A):
  def print(self):
    super().print()
    print('Class - B')

b = B()
b.print()
```

```python
class A:
  def print(self):
    print('Class - A')

class B:
  def print(self):
    print('Class - B')

class C(A, B):
  def print(self):
    A.print(self)
    B.print(self)
    print('Class - C')

c = C()
c.print()
```
