# Patterns

```python
n = int(input("Enter the size: "))
for i in range(n):
  for j in range(i + 1):
    print((j + 1), end = " ")
  print()
```

```python
n = int(input("Enter the size: "))
c = 1
for i in range(n):
  for j in range(i + 1):
    print(c, end = ' ')
    c += 1
  print()
```

```python
n = int(input("Enter the size: "))
for i in range(n):
  for j in range(i + 1):
    print((i * j), end = " ")
  print()
```

```python
n = int(input("Enter the size: "))
for i in range(n):
  for j in range(i + 1):
    print(chr(65 + j), end = ' ')
  print()
```

```python
n = int(input("Enter the size: "))
c = 0
for i in range(n):
  for j in range(i + 1):
    print(chr(65 + c), end = ' ')
    c += 1
  print()
```

# While Loop

```python
l = [12, 20, 36, 42, 55, 67, 74, 83, 98, 101]
for i in range(len(l)):
  if l[i] == 55:
    print(f"Element found at index: {i}")
```

```python
l = [12, 20, 36, 42, 55, 67, 74, 83, 98, 101]
i = 0
while (i < len(l)):
  if l[i] == 55:
    print(f'Element found at index: {i}')
    break
  i += 1
else:
  print('Not Found')
```

```python
l = [120, 220, 36, 42, 545, 67, 724, 83, 98, 101]
i = 0
max_val = 0
while (i < len(l)):
  if (l[i] > max_val):
    max_val = l[i]
  i += 1
print(max_val)
```

```python
l = [120, 220, 36, 42, 545, 67, 724, 83, 98, 101]
i = 0
j = len(l) - 1
while (i <= j):
  if i == j:
    print(l[i])
  else:
    print(l[i] + l[j])
  i += 1
  j -= 1
```

# List Comprehension

```python
l = [1, 2, 3, 4, 5, 56, 6, 7]
res = []
for i in l:
  res.append(i + 10)
print(res)
```

```python
l = [1, 2, 3, 4, 5, 56, 6, 7]
result = [(i + 10) for i in l]
print(result)
```

```python
l = [1, 2, 3, 4, 5, 56, 6, 7]
multi = [(i * l[i]) for i in range(len(l))]
print(multi)
```

```python
l = [1, 2, 3, 4, 5, 56, 6, 7]
cube = [(i ** 3) for i in l]
print(cube)
```

```python
l = [1, 55, 67, 89, 12, 2, 3, 4, 5, 56, 6, 7]
even = [i for i in l if ((i % 2) == 0)]
print(even)
```

```python
l = [1, 55, 67, 89, 12, 2, 3, 4, 5, 56, 6, 7]
even_app = [i if ((i % 2) == 0) else 0 for i in l]
print(even_app)
```
